```python

# Задание 2.1 автоград

x = torch.tensor(2.0, requires_grad=True)
y = torch.tensor(3.0, requires_grad=True)
z = torch.tensor(4.0, requires_grad=True)

f = x**2 + y**2 + z**2 + 2*x*y*z
f.backward()

print(x.grad)
print(y.grad)
print(z.grad)
```

```python
# Задание 2.2 mse

x = torch.randn(10)
y_true = torch.randn(10)

w = torch.tensor(0.5, requires_grad=True)
b = torch.tensor(0.1, requires_grad=True)

y_pred = w * x + b
mse = torch.mean((y_pred - y_true) ** 2)
mse.backward()

print(w.grad)
print(b.grad)
```

```python
# Задание 3.1 данные

A = torch.randn(64, 1024, 1024)
B = torch.randn(128, 512, 512)
C = torch.randn(256, 256, 256)

# у меня cuda работает, но перестрахуемся
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
# Задание 3.2 измерение времени

def cpu_time(fn):
    start = time.time()
    fn()
    end = time.time()
    return (end - start) * 1000


def gpu_time(fn):
    start = torch.cuda.Event(enable_timing=True)
    end = torch.cuda.Event(enable_timing=True)

    start.record()
    fn()
    end.record()

    torch.cuda.synchronize()
    return start.elapsed_time(end)
```

```python
# Задание 3.3 сравнение операций

def test(x):

    cpu_matmul = cpu_time(lambda: torch.matmul(x, x))
    cpu_add = cpu_time(lambda: x + x)
    cpu_mul = cpu_time(lambda: x * x)
    cpu_t = cpu_time(lambda: x.T)
    cpu_sum = cpu_time(lambda: torch.sum(x))

    if torch.cuda.is_available():
        x_gpu = x.to("cuda")

        gpu_matmul = gpu_time(lambda: torch.matmul(x_gpu, x_gpu))
        gpu_add = gpu_time(lambda: x_gpu + x_gpu)
        gpu_mul = gpu_time(lambda: x_gpu * x_gpu)
        gpu_t = gpu_time(lambda: x_gpu.T)
        gpu_sum = gpu_time(lambda: torch.sum(x_gpu))
    else:
        gpu_matmul = None
        gpu_add = None
        gpu_mul = None
        gpu_t = None
        gpu_sum = None

    print("CPU:", cpu_matmul, cpu_add, cpu_mul, cpu_t, cpu_sum)
    print("GPU:", gpu_matmul, gpu_add, gpu_mul, gpu_t, gpu_sum)


test(A)
test(B)
test(C)
```
